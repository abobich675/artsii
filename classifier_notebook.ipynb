```python
def create_patches(img_array, p_size): 
    h, w, c = img_array.shape
    h_patches = h // p_size
    w_patches = w // p_size

    img_array = img_array[:h_patches * p_size, :w_patches * p_size]

    patches = (
        img_array
        .reshape(h_patches, p_size, w_patches, p_size, c)
        .swapaxes(1, 2)  # swap patch grid order
    )
    #patches = image.extract_patches_2d(img_array, (p_size, p_size))
        #print(patches[1])
        #print(f"Shape: {patches.shape}")

    return patches
```

```python
def luminance(patch):
    #print(f'RGB values of one pixel: {patch[0, 0]}')

    R = (patch[0, 0][0])/255
    if R <= 0.0405:
        R = R/12.92
    elif R > 0.0404:
        R = ((R + 0.055)/1.055)**2.4
    G = (patch[0, 0][1])/255
    if G <= 0.0405:
        G = G/12.92
    elif G > 0.0404:
        G = ((G + 0.055)/1.055)**2.4
    B = (patch[0, 0][2])/255
    if B <= 0.0405:
        B = B/12.92
    elif B > 0.0404:
        B = ((B + 0.055)/1.055)**2.4

    L = 0.2126 * R + 0.7152 * G + 0.0722 * B

    return L

    # min_bright = patch.min(),
    # max_bright = patch.max(),
    # contrast = patch.std(),
    # avg_bright = patch.avg()
    # features = [min_bright, max_bright, contrast, avg_bright]
    # print(features)
    # return features
```

```python
def brightness_matching(lum):
    brightness = lum * 59
    brightness = int(brightness)
    ascii_vals = """▓B%8&WM▒#*oahkqmZO0QLCJUYXzcxrjft░\|)1{}[]?-_+~>i!lI;:,"^`' """
    symbol = ascii_vals[brightness]
    return symbol

    #char = symbols[patch_features[3]  69]
```

```python
def col_matching(color):
    saturation = color * 59
    saturation = int(saturation)
    ascii_vals = """▓B%8&WM▒#*oahkqmZO0QLCJUYXzcxrjft░\|)1{}[]?-_+~>i!lI;:,"^`' """
    symbol = ascii_vals[saturation]
    return symbol

def multilayered():
    p_size = 10
    with Image.open("/Users/teganmyers/Downloads/mountain_test.jpg") as im:
        test = np.array(im)
    patches = create_patches(test, p_size)
    h_patches, w_patches = patches.shape[:2]

    red_grid = ""
    green_grid = ""
    blue_grid = ""

    for h in range(h_patches):
        for w in range(w_patches):
            patch = patches[h,w]
            r = (patch[0, 0][0])/255
            g = (patch[0, 0][1])/255
            b = (patch[0, 0][2])/255

            red = col_matching(r)
            green = col_matching(g)
            blue = col_matching(g)
            red_grid += (f"\033[38;2;{255};{0};{0}m{red}\033[0m")
            green_grid += (f"\033[38;2;{0};{255};{0}m{green}\033[0m")
            blue_grid += (f"\033[38;2;{0};{0};{255}m{blue}\033[0m")
        red_grid += '\n'
        green_grid += '\n'
        blue_grid += '\n'

    print(red_grid)
    print(green_grid)
    print(blue_grid)


multilayered()
    
```
